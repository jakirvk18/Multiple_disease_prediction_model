Handling missing values and converting the raw data into required format

```python
import pandas as pd
import numpy as np  
import sys
import os
```

Diabetes data set processing

```python
diabetes_df = pd.read_csv('./data/diabetes.csv')
diabetes_df.head()
```

```python

#change column name of 'Outcome' to 'target'
diabetes_df = diabetes_df.rename(columns={'Outcome': 'target'})
```

```python
diabetes_df.head()
```

```python
#checknig for null values in eveery column
null_values = diabetes_df.isnull().sum()
print("Null values in each column:")
print(null_values)
```

```python
#checking for null values in every column
null_values = heart_df.isnull().sum()
print("Null values in each column:")
print(null_values)
```

Lung cancer dataset preprocessing

```python
lung_df = pd.read_csv('./data/lung_cancer.csv')

#checking for null values in every column
null_values = lung_df.isnull().sum()
print("Null values in each column:")
print(null_values)
```

```python
lung_df.head()
```

```python
lung_df['SMOKING'] = lung_df['SMOKING'].replace({2: 'Yes', 1: 'No'})
lung_df['YELLOW_FINGERS'] = lung_df['YELLOW_FINGERS'].replace({2: 'Yes', 1: 'No'})
lung_df['ANXIETY'] = lung_df['ANXIETY'].replace({2: 'Yes', 1: 'No'})
lung_df['PEER_PRESSURE'] = lung_df['PEER_PRESSURE'].replace({2: 'Yes', 1: 'No'})
lung_df['CHRONIC DISEASE'] = lung_df['CHRONIC DISEASE'].replace({2: 'Yes', 1: 'No'})
lung_df['FATIGUE '] = lung_df['FATIGUE '].replace({2: 'Yes', 1: 'No'})
lung_df['ALLERGY '] = lung_df['ALLERGY '].replace({2: 'Yes', 1: 'No'})
lung_df['WHEEZING'] = lung_df['WHEEZING'].replace({2: 'Yes', 1: 'No'})
lung_df['ALCOHOL CONSUMING'] = lung_df['ALCOHOL CONSUMING'].replace({2: 'Yes', 1: 'No'})
lung_df['COUGHING'] = lung_df['COUGHING'].replace({2: 'Yes', 1: 'No'})
lung_df['SHORTNESS OF BREATH'] = lung_df['SHORTNESS OF BREATH'].replace({2: 'Yes', 1: 'No'})
lung_df['SWALLOWING DIFFICULTY'] = lung_df['SWALLOWING DIFFICULTY'].replace({2: 'Yes', 1: 'No'})
lung_df['CHEST PAIN'] = lung_df['CHEST PAIN'].replace({2: 'Yes', 1: 'No'})
```

```python
lung_df.head()
```

```python
#change col name of 'LUNG_CANCER' to 'target' and fatigue to 'FATIGUE'
lung_df = lung_df.rename(columns={'LUNG_CANCER': 'target', 'FATIGUE ': 'FATIGUE' , 'ALLERGY ': 'ALLERGY'})
```

```python
lung_df.head()
```

```python
#replace 'target' column values with yes no to 0 and 1

lung_df['target'] = lung_df['target'].replace({'NO': 0, 'YES': 1})
```

```python
lung_df.head()
```

```python
# Ensure the output directory exists
os.makedirs('./data_preprocessed', exist_ok=True)

#save the dfs
lung_df.to_csv('./data_preprocessed/lung_cancer_data_processed.csv', index=False)
diabetes_df.to_csv('./data_preprocessed/diabetes_data_processed.csv', index=False)
heart_df.to_csv('./data_preprocessed/heart_disease_data_processed.csv', index=False)
print("Data processing complete. Processed files saved as CSV.")
```
